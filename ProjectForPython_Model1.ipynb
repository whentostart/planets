```python
import numpy as np
import datetime as dt
import pandas as pd
```

```python
ratio2018_1 = pd.read_excel('2018_Pay_Ratio.xlsx', parse_dates=['FilingDate'])
ratio2018_1.head()
```

```python
ratio2018_1.info()
```

```python
bmc_1=pd.read_csv('bmc.csv', parse_dates=['datadate'])
bmc_1.head()
```

```python
m1=pd.merge(ratio2018_1, bmc_1, how='left', on=['Ticker', 'FiscalYear'])
m1.head(10)
```

```python
m1.info()
```

```python
car_1=pd.read_csv('car.csv', parse_dates=['date'])
car_1.head()
```

```python
car_1.info()
```

```python
from datetime import timedelta
m1['begdate']=m1['FilingDate']-timedelta(days=1)
m1['enddate']=m1['FilingDate']+timedelta(days=5)
m1.head()
```

```python
m1.info()
```

```python
import sqlite3
sqlconn = sqlite3.connect(':memory:')
m1.to_sql('m1',sqlconn)
car_1.to_sql('car_1', sqlconn)

qry = '''
     SELECT
    a.Ticker, a.FilingDate, a.PayRatio, a.MktCap, a.BookToMarket, a.begdate, a.enddate, b.date, b.AbnormalReturn
    FROM m1 as a 
    LEFT JOIN car_1 as b
    ON a.Ticker = b.Ticker
    AND a.begdate <= b.date 
    AND b.date <= a.enddate
    ORDER BY No
    '''
```

```python
m2 = pd.read_sql_query(qry, sqlconn)
```

```python
m2.info()
```

```python
m2.head()
```

```python
m2.info()
```

```python
m2=m2.dropna()
m2.shape
```

```python
m2.head(20)
```

```python
m2.describe()
```

```python
m3=m2.copy()
```

```python
import sqlite3
sqlconn = sqlite3.connect(':memory:')
m3.to_sql('m3',sqlconn)


qry = '''
     SELECT
    Ticker, FilingDate, PayRatio, MktCap, BookToMarket, begdate, enddate, date, sum(AbnormalReturn) AbnormalReturn
    FROM m3
    
 
    GROUP BY Ticker
    '''
```

```python
m3 = pd.read_sql_query(qry, sqlconn)
```

```python
m3['lnPayRatio'] = np.log(m3['PayRatio'])
m3['lnMktCap'] = np.log(m3['MktCap'])
```

```python
m3.head()
```

```python
import statsmodels.formula.api as sm
```

```python
#specify the model
model1 = sm.ols(formula='AbnormalReturn~lnPayRatio', data=m3)
#run the model and save results as 'res1'
res1 = model1.fit()
#print the results from the model
res1.summary()
```

```python
m3
```

```python
m3.info()
```

```python
import matplotlib.pyplot as plt

plt.title('Scatter Plot of CEO Pay Ratio (x) and Stock Price (y)')
plt.scatter(m3['lnPayRatio'], m3['AbnormalReturn'])
plt.show()
```

```python
#specify the mod
mode2 = sm.ols(formula='AbnormalReturn~lnPayRatio+lnMktCap+BookToMarket', data=m3)
#run the model
res2 = mode2.fit()
#print out the results
res2.summary()
```

```python
from statsmodels.iolib.summary2 import summary_col
output = summary_col([res1, res2], stars=True)
output
```
